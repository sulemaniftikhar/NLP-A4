```python
import os
import shutil
import json
import torch
import warnings
import chromadb
import ipywidgets as widgets
from IPython.display import display, clear_output
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig

warnings.filterwarnings('ignore')

DATASET_PATH = ""
for root, dirs, files in os.walk("/kaggle/input"):
    for file in files:
        if file.endswith(".json"):
            if "Finished" in root:
                DATASET_PATH = root.split("/Finished")[0] + "/Finished"
                break
    if DATASET_PATH: break

if not DATASET_PATH:
    DATASET_PATH = "/kaggle/input/npl-assignment-4/mimic-iv-ext-direct-1.0.0/Finished"

DB_PATH = "./chroma_db_native"
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
def load_data(root_path):
    documents = []
    seen_ids = set()
    
    if not os.path.exists(root_path):
        return []

    for dirpath, _, filenames in os.walk(root_path):
        for file in filenames:
            if file.lower().endswith('.json'):
                try:
                    full_path = os.path.join(dirpath, file)
                    category = os.path.basename(os.path.dirname(full_path))
                    unique_id = f"{category}_{file}"
                    
                    if unique_id in seen_ids:
                        continue
                    seen_ids.add(unique_id)
                    
                    with open(full_path, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                    
                    text = json.dumps(data) if isinstance(data, (dict, list)) else str(data)
                    documents.append({"id": unique_id, "text": text, "source": file, "category": category})
                except:
                    continue
    return documents

docs = load_data(DATASET_PATH)
print(len(docs))
```

```python
def recursive_split(text, chunk_size=1000, overlap=200):
    chunks = []
    start = 0
    while start < len(text):
        end = min(start + chunk_size, len(text))
        chunks.append(text[start:end])
        if end == len(text): break
        start += (chunk_size - overlap)
    return chunks

embedding_model = SentenceTransformer('all-MiniLM-L6-v2', device=device)

chunked_texts = []
chunk_metadatas = []
chunk_ids = []

for doc in docs:
    splits = recursive_split(doc['text'])
    for i, split in enumerate(splits):
        chunked_texts.append(split)
        chunk_metadatas.append({"source": doc['source'], "category": doc['category']})
        chunk_ids.append(f"{doc['id']}_{i}")

if os.path.exists(DB_PATH):
    shutil.rmtree(DB_PATH)

os.makedirs(DB_PATH, exist_ok=True)

client = chromadb.PersistentClient(path=DB_PATH)
collection = client.create_collection(name="clinical_notes")

batch_size = 200
for i in range(0, len(chunked_texts), batch_size):
    end = min(i + batch_size, len(chunked_texts))
    batch_texts = chunked_texts[i:end]
    batch_ids = chunk_ids[i:end]
    batch_meta = chunk_metadatas[i:end]
    
    embeddings = embedding_model.encode(batch_texts).tolist()
    collection.add(documents=batch_texts, embeddings=embeddings, metadatas=batch_meta, ids=batch_ids)

print(len(chunked_texts))
```

```python
bnb_config = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_use_double_quant=True, bnb_4bit_quant_type="nf4", bnb_4bit_compute_dtype=torch.float16)
model_id = "HuggingFaceH4/zephyr-7b-beta"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config, device_map="auto")
text_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=512, temperature=0.2, return_full_text=False)
```

```python
def get_answer(query):
    q_emb = embedding_model.encode([query]).tolist()
    results = collection.query(query_embeddings=q_emb, n_results=3)
    
    if results['documents']:
        context = "\n---\n".join(results['documents'][0])
    else:
        context = "No info found."
        
    prompt = f"<|system|>Expert Doctor. Answer using context only.<|user|>Context:\n{context}\n\nQuestion: {query}<|assistant|>"
    response = text_pipeline(prompt)[0]['generated_text']
    return response, context

print("RAG Ready")
```

```python
import re

def evaluate_cohesion(query, answer):
    eval_prompt = f"<|system|>Rate Cohesion 1-5 only. OUTPUT ONLY THE NUMBER.<|user|>Q:{query}\nA:{answer}<|assistant|>"
    raw_score = text_pipeline(eval_prompt, max_new_tokens=10)[0]['generated_text']
    
    match = re.search(r'[1-5]', raw_score)
    if match:
        return match.group(0)
    else:
        return "3"
```

```python
test_q = "What is the primary diagnosis?"
ans, ctx = get_answer(test_q)
score = evaluate_cohesion(test_q, ans)

print(test_q)

print(ans)

print(f"Cohesion Score: {score}/5")
```

```python
%%writefile app.py
import streamlit as st
import torch
import chromadb
import os
import re
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig

st.set_page_config(layout="wide", page_title="DiReCT RAG")

@st.cache_resource
def load_resources():
    db_path = "./chroma_db_native"
    if not os.path.exists(db_path):
        return None, None, None
        
    client = chromadb.PersistentClient(path=db_path)
    collection = client.get_collection("clinical_notes")
    
    embed_model = SentenceTransformer('all-MiniLM-L6-v2', device="cuda" if torch.cuda.is_available() else "cpu")
    
    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True, bnb_4bit_use_double_quant=True, 
        bnb_4bit_quant_type="nf4", bnb_4bit_compute_dtype=torch.float16
    )
    model_id = "HuggingFaceH4/zephyr-7b-beta"
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config, device_map="auto")
    
    pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=512, temperature=0.2, return_full_text=False)
    return collection, embed_model, pipe

def evaluate_cohesion_ui(pipe, query, answer):
    prompt = f"<|system|>Rate Cohesion 1-5 only. OUTPUT ONLY THE NUMBER.<|user|>Q:{query}\nA:{answer}<|assistant|>"
    raw_score = pipe(prompt, max_new_tokens=10)[0]['generated_text']
    match = re.search(r'[1-5]', raw_score)
    if match:
        return match.group(0)
    else:
        return "3"

st.title("DiReCT: Clinical Assistant")

try:
    collection, embed_model, pipe = load_resources()
    
    if collection:
        with st.form("main_form"):
            query = st.text_area("Enter Clinical Query:", height=100)
            submitted = st.form_submit_button("Generate", type="primary")
            
            if submitted and query:
                with st.spinner("Processing..."):
                    q_emb = embed_model.encode([query]).tolist()
                    res = collection.query(query_embeddings=q_emb, n_results=3)
                    
                    if res['documents']:
                        docs = res['documents'][0]
                        metas = res['metadatas'][0]
                        context = "\n---\n".join(docs)
                    else:
                        docs = []
                        context = "No info."
                    
                    prompt = f"<|system|>Expert Doctor.<|user|>Context:\n{context}\n\nQuestion: {query}<|assistant|>"
                    ans = pipe(prompt)[0]['generated_text']
                    
                    cohesion = evaluate_cohesion_ui(pipe, query, ans)
                    
                    st.success("Complete")
                    st.subheader("Assessment")
                    st.write(ans)
                    st.info(f"Cohesion Score: {cohesion}/5")
                    
                    if docs:
                        with st.expander("Evidence"):
                            for i, d in enumerate(docs):
                                st.caption(metas[i]['source'])
                                st.code(d[:300])
    else:
        st.error("Database not found.")

except Exception as e:
    st.error(str(e))
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

print("UI Refreshing...")

header = widgets.HTML("<h2>DiReCT: Clinical Diagnostic Assistant</h2>")

query_box = widgets.Textarea(
    value="",
    placeholder='Enter clinical query here (e.g., Treatment for Pneumonia?)',
    layout=widgets.Layout(width='100%', height='100px')
)

btn = widgets.Button(
    description='Generate Diagnosis',
    button_style='primary',
    layout=widgets.Layout(width='200px')
)

clear_btn = widgets.Button(
    description='Clear Output',
    button_style='warning',
    layout=widgets.Layout(width='100px')
)

output_area = widgets.Output()

def run_analysis(b):
    with output_area:
        clear_output()
        
        query = query_box.value
        if not query:
            print("Please enter a query first.")
            return
        
        print(f"Processing: '{query}'...")
        print("Retrieving records & Analyzing with LLM... (Wait ~30s)")
        
        try:
            q_emb = embedding_model.encode([query]).tolist()
            res = collection.query(query_embeddings=q_emb, n_results=3)
            
            if res['documents']:
                docs = res['documents'][0]
                metas = res['metadatas'][0]
                context = "\n---\n".join(docs)
            else:
                docs = []
                context = "No relevant records found."
            
            prompt = f"<|system|>Expert Doctor.<|user|>Context:\n{context}\n\nQuestion: {query}<|assistant|>"
            ans = text_pipeline(prompt)[0]['generated_text']
            
            clear_output()
            
            print("\nASSESSMENT:\n" + "="*60)
            print(ans)
            
            if docs:
                print("\nEVIDENCE SOURCES:\n" + "-"*60)
                for i, d in enumerate(docs):
                    source_name = metas[i].get('source', 'Unknown')
                    category = metas[i].get('category', 'N/A')
                    print(f"[{i+1}] {source_name} ({category})")
                    print(f"{d[:250]}...\n")
                    
        except Exception as e:
            clear_output()
            print(f"Error: {e}")

def clear_screen(b):
    with output_area:
        clear_output()
        print("Ready for new query.")

btn.on_click(run_analysis)
clear_btn.on_click(clear_screen)

ui = widgets.VBox([header, query_box, widgets.HBox([btn, clear_btn]), output_area])
display(ui)
```
